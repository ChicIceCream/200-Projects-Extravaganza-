```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")

# Pre-processes the text, tokenize it, then post process it
result = classifier("I have been waiting for this for a long time! It is finally here!")

print(result)
```

```python
generator = pipeline("text-generation", model = "distilgpt2")

res = generator(
    "In this research paper about AI, we will talk about ",
    max_length = 30,
    num_return_sequences = 5,
    truncation = True
)

print(res)
```

```python
classifier = pipeline("zero-shot-classification")

result = classifier(
    "My bear is cute!",
    candidate_labels = ["animal", "word"]
)

print(result)
```
